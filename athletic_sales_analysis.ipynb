```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
year2020_sales_df = pd.read_csv('Resources/athletic_sales_2020.csv')
year2021_sales_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
year2020_sales_df.head(10)
```

```python
# Display the 2021 sales DataFrame
year2021_sales_df.head(10)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
#year2020_sales_df.dtypes
year2020_sales_df.info()
```

```python
# Check the 2021 sales data types.
year2021_sales_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
year_2020_2021_combined_df = pd.concat([year2020_sales_df, year2021_sales_df], axis='rows')
year_2020_2021_combined_df = year_2020_2021_combined_df.reset_index(drop=True)
year_2020_2021_combined_df
```

```python
# Check if any values are null.
year_2020_2021_combined_df.isnull().sum()
```

```python
# Check the data type of each column
year_2020_2021_combined_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
year_2020_2021_combined_df["invoice_date"] = pd.to_datetime(year_2020_2021_combined_df["invoice_date"], format='%Y-%m-%d')
year_2020_2021_combined_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
year_2020_2021_combined_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
productssoldbyregion = year_2020_2021_combined_df.groupby(["region", "state", "city"])["units_sold"].sum().reset_index(name="Total_Products_Sold").sort_values(by="Total_Products_Sold", ascending=False)
# Show the top 5 results.
productssoldbyregion.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivotProducts = pd.pivot_table(year_2020_2021_combined_df, values="units_sold", index=["region", "state", "city"], aggfunc="sum", fill_value=0).sort_values(by="units_sold", ascending=False)
# Rename the "units_sold" column to "Total_Products_Sold"
pivotProducts = pivotProducts.rename(columns={"units_sold": "Total_Products_Sold"})
# Show the top 5 results.
pivotProducts.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
mostsalesbyregion = year_2020_2021_combined_df.groupby(["region", "state", "city"])["total_sales"].sum().to_frame().sort_values(by="total_sales", ascending=False)
mostsalesbyregion = mostsalesbyregion.rename(columns={"total_sales": "Total_Sales"})
# Show the top 5 results.
mostsalesbyregion.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivotTotalSales = pd.pivot_table(year_2020_2021_combined_df, values="total_sales", index=["region", "state", "city"], aggfunc="sum", fill_value=0).sort_values(by="total_sales", ascending=False)
# Optional: Rename the "total_sales" column to "Total Sales"
pivotTotalSales = pivotTotalSales.rename(columns={"total_sales": "Total_Sales"})
# Show the top 5 results.
pivotTotalSales.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
retailerSales = year_2020_2021_combined_df.groupby(["retailer", "region", "state", "city"])["total_sales"].sum().to_frame().sort_values(by="total_sales", ascending=False)
retailerSales = retailerSales.rename(columns={"total_sales": "Total_Sales"})
# Show the top 5 results.
retailerSales.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womensfootwear = year_2020_2021_combined_df[year_2020_2021_combined_df["product"].str.contains("Women's Athletic Footwear")]
womensfootwear.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womensfootweargroup = womensfootwear.groupby(["retailer", "region", "state", "city"])["units_sold"].sum().to_frame().sort_values("units_sold", ascending=False)
womensfootweargroup = womensfootweargroup.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})
# Show the top 5 results.
womensfootweargroup.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivotwomensfootweargroup = pd.pivot_table(womensfootwear, values="units_sold", index=["retailer", "region", "state", "city"], aggfunc="sum", fill_value=0).sort_values(by="units_sold", ascending=False)
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivotwomensfootweargroup = pivotwomensfootweargroup.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})
# Show the top 5 results.
pivotwomensfootweargroup.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivotmostsalesday = pd.pivot_table(womensfootwear, values="total_sales", index="invoice_date", aggfunc="sum", fill_value=0)
# Optional: Rename the "total_sales" column to "Total Sales"
pivotmostsalesday = pivotmostsalesday.rename(columns={"total_sales": "Total_Sales"})
# Show the table.
pivotmostsalesday.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
dailysalesbin = pivotmostsalesday.resample("D").sum()
# Sort the resampled pivot table in descending order on "Total Sales".
dailysalesbin = dailysalesbin.sort_values(by="Total_Sales", ascending=False)
dailysalesbin
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weeklysalesbin = pivotmostsalesday.resample("W").sum()
# Sort the resampled pivot table in descending order on "Total Sales".
weeklysalesbin = weeklysalesbin.sort_values(by="Total_Sales", ascending=False)
weeklysalesbin
```
